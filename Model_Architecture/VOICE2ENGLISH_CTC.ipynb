```python
# Custom collate_fn for DataLoader to handle padding
def collate_fn(batch):
    spectrograms, labels = zip(*batch)

    # Pad spectrograms
    max_spec_len = max(s.shape[0] for s in spectrograms)
    padded_spectrograms = torch.zeros(len(spectrograms), max_spec_len, spectrograms[0].shape[1])
    for i, spec in enumerate(spectrograms):
        padded_spectrograms[i, :spec.shape[0], :] = spec

    # Pad labels
    max_label_len = max(l.shape[0] for l in labels)
    padded_labels = torch.zeros(len(labels), max_label_len, dtype=torch.long)
    for i, lab in enumerate(labels):
        padded_labels[i, :lab.shape[0]] = lab

    spectrogram_lengths = torch.tensor([s.shape[0] for s in spectrograms], dtype=torch.long)
    label_lengths = torch.tensor([l.shape[0] for l in labels], dtype=torch.long)

    return padded_spectrograms, padded_labels, spectrogram_lengths, label_lengths

print("\n--- DEBUG: `collate_fn` defined for DataLoader. ---")
```

```python
class ASRDataset(Dataset):
    def __init__(self, dataset, char_to_num):
        self.dataset = dataset
        self.char_to_num = char_to_num

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        try:
            spectrogram = self.dataset[idx]["spectrogram"]
            label = self.dataset[idx]["label"]
            # Ensure spectrogram and label are in a format that can be converted to tensors
            if not isinstance(spectrogram, np.ndarray):
                spectrogram = np.array(spectrogram)
            if not isinstance(label, list): # Assuming label is a list of integers
                 # Attempt to convert label to list if it's not already
                 label = list(label)

            return torch.tensor(spectrogram, dtype=torch.float32), torch.tensor(label, dtype=torch.long)
        except Exception as e:
            print(f"--- DEBUG: Skipping sample {idx} due to error: {e} ---")
            return None

print("DONE")
```

```python
# 7. DEFINE THE CTC MODEL IN PYTORCH
import torch
import torch.nn as nn

class ASRModel(nn.Module):
    def __init__(self, input_dim, output_dim, rnn_layers=2, rnn_units=128, conv_channels=16):
        super().__init__()

        # Define kernel sizes and strides
        kernel_size1 = (11, 41)
        stride1 = (2, 2)
        kernel_size2 = (11, 21)
        stride2 = (1, 2)  # This stride does not affect the time dimension

        # Calculate padding for Conv1 and Conv2
        padding1 = ((kernel_size1[0] - 1) // 2, (kernel_size1[1] - 1) // 2)
        padding2 = ((kernel_size2[0] - 1) // 2, (kernel_size2[1] - 1) // 2)

        # Convolutional layers
        self.conv1 = nn.Sequential(
            nn.Conv2d(1, conv_channels, kernel_size=kernel_size1, stride=stride1, padding=padding1, bias=False),
            nn.BatchNorm2d(conv_channels),
            nn.ReLU()
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(conv_channels, conv_channels, kernel_size=kernel_size2, stride=stride2, padding=padding2, bias=False),
            nn.BatchNorm2d(conv_channels),
            nn.ReLU()
        )

        # Calculate RNN input size after convolutions
        with torch.no_grad():
            dummy_input = torch.randn(1, 1, 100, input_dim)  # (batch, channels, time, features)
            dummy_output = self.conv1(dummy_input)
            dummy_output = self.conv2(dummy_output)
            rnn_input_features = dummy_output.shape[1] * dummy_output.shape[3]  # channels * freq_dim

        # Stacked BiLSTMs
        self.rnns = nn.ModuleList()
        for i in range(rnn_layers):
            self.rnns.append(
                nn.Sequential(
                    nn.LSTM(
                        input_size=rnn_input_features if i == 0 else rnn_units * 2,
                        hidden_size=rnn_units,
                        bidirectional=True,
                        batch_first=True
                    ),
                    nn.Dropout(0.5) if i < rnn_layers - 1 else nn.Identity()
                )
            )

        # Dense + Classifier
        self.dense1 = nn.Sequential(
            nn.Linear(rnn_units * 2, rnn_units),  # 256 → 128
            nn.ReLU(),
            nn.Dropout(0.5)
        )
        self.classifier = nn.Linear(rnn_units, output_dim)  # 128 → vocab_size

    def forward(self, x):
        x = x.unsqueeze(1)  # Add channel dimension (batch, 1, time, input_dim)
        x = self.conv1(x)
        x = self.conv2(x)

        # Reshape for RNN: (batch, channels, time, features) -> (batch, time, channels * features)
        batch_size, channels, time, features = x.shape
        x = x.permute(0, 2, 1, 3).contiguous().view(batch_size, time, channels * features)

        for i, rnn_layer in enumerate(self.rnns):
            x, _ = rnn_layer[0](x)  # LSTM output
            if isinstance(rnn_layer[1], nn.Dropout):
                x = rnn_layer[1](x)

        x = self.dense1(x)
        x = self.classifier(x)
        return x


# Example usage
input_dim = 256 // 2 + 1  # From spectrogram n_fft
output_dim = len(char_to_num)  # Includes blank token
rnn_units = 128
rnn_layers = 2
conv_channels = 32

device = "cuda" if torch.cuda.is_available() else "cpu"
model = ASRModel(input_dim, output_dim, rnn_layers=rnn_layers, rnn_units=rnn_units, conv_channels=conv_channels).to(device)


# Function to count model parameters
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

print("\n\n" + "="*80)
print("--- DEBUG: MODEL SUMMARY ---")
print(model)
print(f"Total trainable parameters: {count_parameters(model)}")
print("="*80 + "\n")
```

```python
# 8. TRAINING AND EVALUATING
# A utility function to decode the output of the network
def decode_batch_predictions(output_logits, input_lengths, num_to_char, blank_idx=0):
    # output_logits: (batch, time, num_classes)
    # input_lengths: (batch,)

    # Get the index of the most probable character at each timestep
    predicted_ids = torch.argmax(output_logits, dim=2) # (batch, time)

    output_text = []
    for i in range(predicted_ids.shape[0]): # Iterate over batch
        current_prediction = predicted_ids[i, :input_lengths[i]].tolist() # Slice to actual length and convert to list

        # CTC greedy decoding rules:
        # 1. Remove consecutive duplicates
        # 2. Remove blank tokens
        decoded_tokens = []
        for token_id in current_prediction:
            if token_id == blank_idx:
                continue
            if len(decoded_tokens) == 0 or token_id != decoded_tokens[-1]:
                decoded_tokens.append(token_id)

        # Convert numeric tokens back to characters
        text = "".join([num_to_char[token_id] for token_id in decoded_tokens])
        output_text.append(text)
    return output_text


print("\n--- DEBUG: `decode_batch_predictions` function defined. ---")


class CallbackEval:
    def __init__(self, model, dataloader, num_to_char, device, time_reduction_factor):
        self.model = model
        self.dataloader = dataloader
        self.num_to_char = num_to_char
        self.device = device
        self.time_reduction_factor = time_reduction_factor

    def on_epoch_end(self, epoch: int, logs=None):
        self.model.eval() # Set model to evaluation mode
        predictions = []
        targets = []

        # Take a single batch from the validation dataset for a quick check
        for batch_idx, (spectrograms, labels, spec_lens, label_lens) in enumerate(self.dataloader):
            if batch_idx == 0: # Only process the first batch
                spectrograms = spectrograms.to(self.device)

                with torch.no_grad():
                    output_logits = self.model(spectrograms)

                # Adjust spec_lens for the model's output time dimension
                adjusted_spec_lens = torch.div(spec_lens, self.time_reduction_factor, rounding_mode='trunc').to(torch.long)
                # Ensure adjusted_spec_lens is at least 1, as CTC loss requires positive lengths
                adjusted_spec_lens[adjusted_spec_lens == 0] = 1


                batch_predictions = decode_batch_predictions(output_logits, adjusted_spec_lens, self.num_to_char, blank_idx=0) # Pass blank_idx
                predictions.extend(batch_predictions)

                for i in range(labels.shape[0]):
                    label_tensor = labels[i][:label_lens[i]] # Slice to actual length
                    label_str = "".join([self.num_to_char[c.item()] for c in label_tensor])
                    targets.append(label_str)
                break # Process only the first batch

        wer_score = wer(targets, predictions)
        print("-" * 100)
        print(f"Epoch {epoch+1} - Word Error Rate (on one validation batch): {wer_score:.4f}")
        print("-" * 100)

        for i in np.random.randint(0, len(predictions), 2):
            print(f"Target    : {targets[i]}")
            print(f"Prediction: {predictions[i]}")
            print("-" * 100)
        self.model.train() # Set model back to training mode


print("\n--- DEBUG: `CallbackEval` class defined. ---")
```

```python
# Define the number of epochs.
epochs = 27

criterion = nn.CTCLoss(blank=0, zero_infinity=True).to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-4)

# Calculate the time reduction factor due to convolutions
# Stride in time for conv1 is 2
# Stride in time for conv2 is 1
# Total time reduction factor = 2 * 1 = 2
time_reduction_factor = 2

validation_callback = CallbackEval(model, validation_loader, num_to_char, device, time_reduction_factor)

print("\n--- DEBUG: Starting model training... ---")

history = {'loss': [], 'val_loss': []} # To store loss values

for epoch in range(epochs):
    model.train()
    running_loss = 0.0

    # --- Training loop ---
    train_iter = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs} - Training", leave=False)
    for batch_idx, (spectrograms, labels, spec_lens, label_lens) in enumerate(train_iter):
        spectrograms = spectrograms.to(device)
        labels = labels.to(device)
        adjusted_spec_lens = torch.div(spec_lens, time_reduction_factor, rounding_mode='trunc').to(torch.long)
        adjusted_spec_lens[adjusted_spec_lens == 0] = 1
        label_lens = label_lens.to(device)

        optimizer.zero_grad()
        output_logits = model(spectrograms)
        log_probs = torch.nn.functional.log_softmax(output_logits, dim=2).permute(1, 0, 2)
        loss = criterion(log_probs, labels, adjusted_spec_lens, label_lens)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        train_iter.set_postfix({"loss": running_loss / (batch_idx + 1)})

    avg_train_loss = running_loss / len(train_loader)
    history['loss'].append(avg_train_loss)

    # --- Validation loop ---
    model.eval()
    val_running_loss = 0.0
    val_iter = tqdm(validation_loader, desc=f"Epoch {epoch+1}/{epochs} - Validation", leave=False)
    with torch.no_grad():
        for batch_idx, (spectrograms, labels, spec_lens, label_lens) in enumerate(val_iter):
            spectrograms = spectrograms.to(device)
            labels = labels.to(device)
            adjusted_spec_lens = torch.div(spec_lens, time_reduction_factor, rounding_mode='trunc').to(torch.long)
            adjusted_spec_lens[adjusted_spec_lens == 0] = 1
            label_lens = label_lens.to(device)

            output_logits = model(spectrograms)
            log_probs = torch.nn.functional.log_softmax(output_logits, dim=2).permute(1, 0, 2)
            loss = criterion(log_probs, labels, adjusted_spec_lens, label_lens)
            val_running_loss += loss.item()
            val_iter.set_postfix({"val_loss": val_running_loss / (batch_idx + 1)})

    avg_val_loss = val_running_loss / len(validation_loader)
    history['val_loss'].append(avg_val_loss)

    print(f"Epoch {epoch+1}/{epochs}, Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}")
    validation_callback.on_epoch_end(epoch)

    checkpoint_path = f"CTC_asr_epoch{epoch}.pt"
    torch.save({
            "epoch": epoch + 1,
            "model_state_dict": model.state_dict(),
            "optimizer_state_dict": optimizer.state_dict(),
            "train_loss": avg_train_loss,
            "val_loss": avg_val_loss,
        }, checkpoint_path)
    print(f"Checkpoint saved: {checkpoint_path}")
```
